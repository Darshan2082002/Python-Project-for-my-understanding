```python
# Numpy and Pytorch tensors
# Numpy is a popular scientific python numerical computing
# and because of this, pytorch has functionality to interact with it
# data in Numpy  want in pytorch tensor 'torch.from_numpy(ndarray) '
# Pytoech tensor-> Numpy-> torch.tensor.numpy()
```

```python
import torch 
import numpy as np 
array=np.arange(1,8)
print(array)
print(array.shape)
print(array.size)
tensor=torch.from_numpy(array).type(torch.float32)# waring: when converting from numpy -> pytorch  pytorch reflect numpy default  one 

print('tensor ')
print(f"Tensor size:{tensor.size}")
print(f"Tensor shape:{tensor.shape}")
print(f"Tensor {tensor.dtype}")
```

Pytorch Reproducibility 


```python
### Reproducbility (trying to take random out of random )
# In short how a neural network learns:
' Start with random numbers-> tensor operation-> update random numbers to try and make them of the data -> again-> again-> again..'
' To reduce the randomness in neural networks and Pytorch comes to concept of a random seed'
' Essentially what the random seed does is flavour the randomness'
```

```python
torch.rand(3,3)
```

```python
import torch
#create two random tensors
random_A=torch.rand(3,4)
random_B=torch.rand(3,4)
print(random_A)
print(random_B)
print(random_A==random_B)
```

```python
1==1
```

```python
# Lets's make some random but reproducible tensors
from networkx import random_clustered_graph
import torch 
# set the random seed
Random_seed=42
torch.manual_seed(Random_seed)
random_c=torch.rand(3,4)
print(random_c)
random_d=torch.rand(3,4)
torch.manual_seed(Random_seed)
print(random_d)
```
