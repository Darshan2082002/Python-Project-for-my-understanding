```python
class simple_env():
    def __init__(self):
        self.start_state=0
        self.goal_state=4
        self.state=self.start_state
    def reset(self):
        self.state=self.start_state
        return self.state
    def step(self,action):
        if action==0:
            self.state=(max(0,self.state-1))
        elif action==1:
            self.state=(min(4,self.state+1))
        if self.state==self.goal_state:
            reward=10
            done=True
        else:
            reward=-1
            done=False
        return self.state,reward,done
    
```

```python
env = simple_env()

state = env.reset()
print("Start state:", state)

state, reward, done = env.step(1)  # move right
print("State:", state, "Reward:", reward)
```

```python
# return total reward
env=simple_env()
state=env.reset()
while True:
    total_reward=0
    state,reward,done=env.step(1)
    total_reward+=reward
    if done:
        break
print("Reward:",total_reward)
```

```python
env=simple_env()
gamma=0.9
state=env.reset()
G=0
t=0
while True:
    action=1
    state,reward,done=env.step(action)
    G+=reward*(gamma**t)
    t+=1
    if done:
        break
print("Total discounted reward:",G)
```
