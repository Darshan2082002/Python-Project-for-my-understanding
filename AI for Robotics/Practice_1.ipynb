```python
class simple_env():
    def __init__(self):
        self.start_state=0
        self.goal_state=4
        self.state=self.start_state
    def reset(self):
        self.state=self.start_state
        return self.state
    def step(self,action):
        if action==0:
            self.state=(max(0,self.state-1))
        elif action==1:
            self.state=(min(4,self.state+1))
        if self.state==self.goal_state:
            reward=10
            done=True
        else:
            reward=-1
            done=False
        return self.state,reward,done
    
```

```python
env = simple_env()

state = env.reset()
print("Start state:", state)

state, reward, done = env.step(1)  # move right
print("State:", state, "Reward:", reward)
```

```python
# return total reward
env=simple_env()
state=env.reset()
while True:
    total_reward=0
    state,reward,done=env.step(1)
    total_reward+=reward
    if done:
        break
print("Reward:",total_reward)
```

```python
env=simple_env()
gamma=0.9
state=env.reset()
G=0
t=0
while True:
    action=1
    state,reward,done=env.step(action)
    G+=reward*(gamma**t)
    t+=1
    if done:
        break
print("Total discounted reward:",G)
```

```python
import gymnasium as gym 
from gymnasium import spaces 
import numpy as np 
```

```python
class globalenv(gym.Env):
   

    def __init__(self):
        super().__init__()
        self.row=3
        self.col=3
        self.start_state=(0,0)
        self.goal_state=(4,4)
        self.obstacle_state=(1,1)
        self.state=self.start_state
        self.action_space=spaces.Discrete(4)
        self.observation_space=spaces.MultiDiscrete([self.row,self.col])

    def reset(self):
        self.state=self.start_state
        return np.array(self.state),{}
    def step(self,action):
        row,col=self.state
        if action==0:
            row=-1
        elif action==1:
            row+=1
        elif action==2:
            col-=1
        elif action==3:
            col+=1
        
        row=max(0,min(row,self.row-1))
        col=max(0,min(col,self.col-1))

        next_state=(row,col)
        if next_state==self.obstacle_state:
            reward=-5
            terminated=False
            next_state=self.state
        elif next_state==self.goal_state:
            reward=10
            terminated=True
        else:
            reward=-1
            terminated=False
        self.state=next_state
        return np.array(next_state),reward,terminated,False,{}
```

```python
env=globalenv()
state,_=env.reset()
print("Start:", state)

for _ in range(5):
    action = env.action_space.sample()
    state, reward, done,_,_= env.step(action)
    print("State:", state, "Reward:", reward)
    if done:
        break

```
