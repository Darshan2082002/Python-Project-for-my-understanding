```python
## Tribonnaci Number 
def tri(n):
    if (n==0 or n==1 or n==2):
        return 0
    elif (n==3):
        return 1
    else:
        return tri(n-1)+tri(n-2)+tri(n-3)
   

```

```python
tri(10)
```

```python
def printTrib(n) :
    for i in range(1, n+1) :
        print( tri(i) , " ", end = "")
        

# Driver code
n = 10
printTrib(n)
```
