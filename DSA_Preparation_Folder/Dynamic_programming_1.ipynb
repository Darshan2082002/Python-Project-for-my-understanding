## Top to Bottom


```python
def tribonacci(n):
    h={}
    def tri(n):
        if n in h:
            return h[n]
        elif(n==0 or n==1):
            return 0
        elif(n==2):
            return 1
        else:
            res=tri(n-3)+tri(n-2)+tri(n-1)
            h[n]=res
        return res
    return tri(n)
n=10
for i in range(n):
    print(tribonacci(i),end=' ')
            
```

```python
## Table method 

def tri(n):
    dp=[0]*n
    dp[0]=dp[1]=0
    dp[2]=1
    for i in range(3,n):
        dp[i]=dp[i-3]+dp[i-2]+dp[i-1]

    for i in range(0,n):
        print(dp[i],"\n")
n=10
tri(n)
```

```python
###Lucas Numbers

def lucas(n):
    dp=[0]*(n+1)
    dp[0]=2
    dp[1]=1
    for i in range(2,n+1):
        dp[i]=dp[i-1]+dp[i-2]

    return dp
```

```python
lucas(3)
```
