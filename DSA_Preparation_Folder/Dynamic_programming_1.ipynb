## Top to Bottom


```python
def tribonacci(n):
    h={}
    def tri(n):
        if n in h:
            return h[n]
        elif(n==0 or n==1):
            return 0
        elif(n==2):
            return 1
        else:
            res=tri(n-3)+tri(n-2)+tri(n-1)
            h[n]=res
        return res
    return tri(n)
n=10
for i in range(n):
    print(tribonacci(i),end=' ')
            
```
