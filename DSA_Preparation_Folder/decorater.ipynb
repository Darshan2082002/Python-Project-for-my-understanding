```python
def hello(func):
    def inner():
        print("Start")
        func()
    return inner

def h1():
    print("Darshan")

obj=hello(h1)

obj()
```

```python
def who():
    print("Darshan")

def hello(func):
    def inner():
        print(" The current user:",end=" ")
        func()
    return inner

obj=hello(who)
obj()
```

```python
# @is the decorator symbol 

def hello(func):
    def inner():
        print(" The current user:",end=" ")
        func()
    return inner
@hello    
def who():
    print("Darshan")


who()



```
