```python
import math
def perfect_square():
  n,b=map(int,input().split())
  squares_up_to_b=math.isqrt(b)
  print(squares_up_to_b)
  square_before_n=math.isqrt(n-1)
  print(square_before_n)
  count=squares_up_to_b-square_before_n
  print(count)


perfect_square()
```

```python
import multiprocessing
from multiprocessing import Pool
# This function must be defined globally at the top level
def f(x):
    return x * x


if __name__ == "__main__":
    # Keeping the pool size reasonable
    with Pool(processes=3) as p:
        result = p.map(f, [1, 2, 3])
        print(result)
```

```python
def count_no():
    n = input()
    a = list(map(int, n.split()))
    count = 0
    for i in range(len(a)):
        if i >= 0:
            count += 1
    return count

count_no()
```
